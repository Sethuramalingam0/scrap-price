```python
df1.drop(['ID','symboling','fueltypes','aspiration','drivewheels','enginelocation','wheelbase','carlength','carwidth','enginesize','stroke','citympg','highwaympg'],axis=1,inplace = True)
```

```python
df1['doornumbers']=df1['doornumbers'].replace({'two':2,'four':4})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['name'] = le.fit_transform(df1['name'])
df1['carbody'] = le.fit_transform(df1['carbody'])
df1['enginetype']= le.fit_transform(df1['enginetype'])
df1['cylindernumber']=le.fit_transform(df1['cylindernumber'])
df1['fuelsystem']=le.fit_transform(df1['fuelsystem'])
df1
```

```python
x = df1.iloc[:,0:12].values
y = df1.iloc[:,[-1]]
```

```python
x
```

```python
y
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
sd.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.3,random_state = 43)
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict  =lr.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_predict,y_test))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_predict,y_test))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_predict,y_test))
```
